```python
#Figure 1, Complex Plane Visualization
def complex_graph(a):
    import matplotlib.pyplot as plt
    import numpy as np
    for x in range(len(a)):
        plt.plot([0,a[x].real],[0,a[x].imag],'co-',label='python')
    limit=np.max(np.ceil(np.absolute(a))) # set limits for axis
    plt.title('Complex Plane Visual')
    plt.xlim((-1,10))
    plt.ylim((-1,limit))
    plt.ylabel('Imaginary')
    plt.xlabel('Real')
    plt.grid()
    plt.show()
```

```python
b = np.arange(10) + 1j*np.arange(6,16)
complex_graph(b) 
```

```python
#Figure 2, comparing complex numbers and their conjugates, 
#Contour plot
import numpy as np
import matplotlib.pyplot as plt

# making fake data
x = np.linspace(0, 10)
y = np.linspace(0, 10)
c = x[:,np.newaxis] * y
c2 = np.flipud(c)

# plot
fig, ax = plt.subplots(1, 1)
cont1 = plt.contour(x, y, c, colors='maroon',clabel='Complex Number')
cont2 = plt.contour(x, y, c2, colors='salmon',clabel='Complex Conjugate')
cont1.clabel()
cont2.clabel()
plt.title('Contour Plot')
plt.ylabel('Imaginary')
plt.xlabel('Real')
plt.grid()
plt.show()
```

```python
#Figure 4, Fourier series
from __future__ import division
import numpy as np
import pylab as py

# Define "x" range.
x = np.linspace(0, 10, 1000)

# Define "T", i.e functions' period.
T = 2
L = T / 2

# "f(x)" function definition.
def f(x): 
    return np.sin((np.pi) * x) + np.sin((2 * np.pi) * x) + np.sin((5 * np.pi) * x)

# "a" coefficient calculation.
def a(n, L, accuracy = 1000):
    a, b = -L, L
    dx = (b - a) / accuracy
    integration = 0
    for x in np.linspace(a, b, accuracy):
        integration += f(x) * np.cos((n * np.pi * x) / L)
    integration *= dx
    return (1 / L) * integration

# "b" coefficient calculation.
def b(n, L, accuracy = 1000):
    a, b = -L, L
    dx = (b - a) / accuracy
    integration = 0
    for x in np.linspace(a, b, accuracy):
        integration += f(x) * np.sin((n * np.pi * x) / L)
    integration *= dx
    return (1 / L) * integration

# Fourier series.   
def Sf(x, L, n = 10):
    a0 = a(0, L)
    sum = np.zeros(np.size(x))
    for i in np.arange(1, n + 1):
        sum += ((a(i, L) * np.cos((i * np.pi * x) / L)) + (b(i, L) * np.sin((i * np.pi * x) / L)))
    return (a0 / 2) + sum   

# x axis.
py.plot(x, np.zeros(np.size(x)), color = 'black')

# y axis.
py.plot(np.zeros(np.size(x)), x, color = 'black')

# Original signal.
py.plot(x, f(x), linewidth = 1.5, label = 'Signal')

# Approximation signal (Fourier series coefficients).
py.plot(x, Sf(x, L), '.', color = 'palevioletred', linewidth = 1.5, label = 'Fourier series')

# Specify x and y axes limits.
py.xlim([0, 5])
py.ylim([-2.2, 2.2])

py.legend(loc = 'upper right', fontsize = '10')
py.grid()
py.show()
```

```python
#Figure 5, fourier transform visual
x = np.linspace(0,5,100)
y = np.sin(2*np.pi*x)
plt.title('Signal Processing Data')
plt.plot(y, 'darkslateblue')
plt.xlabel('x (in seconds)')
plt.ylabel('Magnitude Value')
plt.grid()
plt.show()

## fourier transform
f = np.fft.fft(y)
## sample frequencies
freq = np.fft.fftfreq(len(y), d=x[1]-x[0])
plt.plot(y)
plt.plot(freq, abs(f)**2, 'darkslateblue') ## will show a peak at a frequency of 1 as it should.
plt.title('Decomposed Data')
plt.ylabel('Square Modulus of Fourier Transform')
plt.xlabel('Frequency (in Hz)')
plt.grid()
plt.show()
```

```python
#Figure 6, Dirac Delta Function
from scipy import signal
signal.unit_impulse((4, 4), 2)
imp = signal.unit_impulse(100, 'mid')
b, a = signal.butter(4, 0.2)
response = signal.lfilter(b, a, imp)
import matplotlib.pyplot as plt
plt.plot(np.arange(-50, 50), imp, 'lightcoral')
plt.plot(np.arange(-50, 50), response, 'orchid')
plt.title('Dirac Delta Function')
plt.margins(0.1, 0.1)
plt.xlabel('Time [samples]')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

```python
#Figure 6, shows linear separability via a nonlinear mapping 
#into a higher-dimensional space
from scipy.interpolate import spline
a = np.arange(-10,10,1)
b = [0] * 20
a_new = np.linspace(a.min(),a.max(),300)
b_smooth = spline(a,b,a_new)
plt.scatter(a_new,b_smooth)
plt.plot(a_new,b_smooth)
plt.grid()
plt.show()
```

```python
c = np.cos(a)
c_smooth = spline(a,c,a_new)
plt.scatter(a_new,c_smooth)
plt.plot(a_new,c_smooth)
plt.grid()
plt.show()
```

```python
#Figure 7, cosine wave for Schrodinger explanation
a = np.arange(-1,10,1)
c = np.cos(a)
c_smooth = spline(a,c,a_new)
plt.xlim((-1,10))
plt.xlabel('x')
plt.ylabel('y')
plt.scatter(a_new,c_smooth)
plt.grid()
plt.show()
```

```python
#Figure 8, Gaussian distribution
lambda_1 = -1
a = 1
x = np.arange(-5,7,1)
A = np.sqrt(3.14/a)
from math import e
rho = A*(e**(lambda_1*(x-a)**2))
x_new = np.linspace(x.min(),x.max(),300)
rho_smooth = spline(x,rho,x_new)
print(x)
print(rho)
plt.scatter(x_new, rho_smooth)
plt.grid()
plt.show()
```
