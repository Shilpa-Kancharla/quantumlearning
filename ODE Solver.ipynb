```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.integrate import odeint
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x = np.arange(-5, 5, 0.25)
y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(x, y)
F = X - Y

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, F)
plt.show()
```

```python
#First order differential equations.
def dydx(y, x):
    return x - y

x0 = np.linspace(0,5,100)
y0 = 1.0  #initial value
y1 = odeint(dydx, y0, x0)
y1 = np.array(y1).flatten()
```

```python
#Plot solution
plt.rcParams.update({'font.size': 14})  # increase the font size
plt.xlabel("x")
plt.ylabel("y")
plt.plot(x0, y1);
```

```python
#Difference in series
plt.xlabel("x")
plt.ylabel("y")
yexact = x0 - 1 + 2*np.exp(-x0)
y_difference = y1 - yexact
plt.plot(x0, y1, x0, yexact, "+");
```

```python
x = np.arange(-5, 5, 0.25)
y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(x, y)
F = X - 1 + 2*np.exp(-X)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, F)
plt.show()
```

```python
#Error calculation
ydiff = np.abs(yexact - y1)
plt.semilogy(x0, ydiff, color="maroon") #logarithmic scale on y-axis
plt.ylabel("Error")
plt.xlabel("x")
plt.title("Error in numerical integration");
```

```python
#Second order differential equations
def d2dx(U, x):
    return [U[1], -2*U[1] - 2*U[0] + np.cos(2*x)]
U0 = [0, 0]
x0 = np.linspace(0, 10, 200)
U1 = odeint(d2dx, U0, x0)
y0 = U1[:,0]
```

```python
plt.xlabel("x")
plt.ylabel("y")
plt.title("Damped harmonic oscillator")
plt.plot(x0,y0, color="navy");
```
